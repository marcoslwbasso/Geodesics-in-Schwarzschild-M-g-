### Órbitas geodésicas no espaço-tempo de Schwarzschild

Dado $(M,g)_{sch}$ onde a métrica é dada por
$$
ds^2 = - f(r) dt^2 + f^{-1}(r) dr^2 + r^2(d\theta^2 + \sin \theta) d\phi^2),
$$
com $f(r) = 1 - 2m/r$. Como tal espaço-tempo é esfericamente simétrica e estático, temos que
\begin{align}
& E:= -g(\partial_t, p) = -p_t,\\
& L:= -g(\partial_{\phi}, p) = p_{\phi}
\end{align}
são constantes de movimento. Ainda, dado $p^{\mu}  = d x^{\mu}/ d \lambda$, então
\begin{align}
 &\dot{t} = f^{-1}(r) E,
 &\dot{\phi} = L/r^2. 
\end{align}

### Geodésicas tipo-tempo

Além disso, dado que $g(p,p) = p_{\mu}p^{\mu} = - \mu^2$, segue que
\begin{align}
\dot{r}^2 - \frac{2m}{r} + \frac{l^2}{r^2}(1 -\frac{2m}{r}) = \epsilon^2 - 1 = \mathcal{E}, 
\end{align}
onde $l = L/\mu$ e $\epsilon = E/\mu$. Ainda, podemos definir o potencial efetivo
\begin{align}
\mathcal{U}_{ef}(r) =  - \frac{m}{r} + \frac{l^2}{2r^2}(1 -\frac{2m}{r}),
\end{align}
de modo a escrever
\begin{align}
\frac{1}{2}\dot{r}^2 + \mathcal{U}_{ef} = \mathcal{E}'.
\end{align}

Os extremos são dados por
\begin{align}
\frac{d \mathcal{U}_{ef}}{dt}|_{r = r_*} = 0 \implies r^{\pm}_* = \frac{1}{2m}\Big(l^2 \pm \sqrt{l^4 - 12l^2m^2},  \Big)
\end{align}
de modo que é necessário que $l \ge l_{c} = 2\sqrt{3}m$ para tais extremos existirem.

Para $m = 10^6 m_*$, onde $m_*$ é a massa solar,  $G = 4 \pi^2 (UA)^3m^{-1}_* yr^{-2}$ e $c \approx 63197.8 UA yr^{-1}$

então
\begin{align}
& 2 \mathcal{U}_{ef}(r) =  - \frac{0.0000463563}{r} + \frac{0.00234489}{r^2} - \frac{0.0197691}{r^3} \\
& r_*^- \approx 0.03474132510043436 \text{UA}, \ \ \ \ r_*^+ \approx 0.20248728156499723 \text{AU}, \ \ \ \ r_s =\frac{2Gm}{c^2} \approx 0.01976905055545263 \text{UA}. 
\end{align}
de modo que
\begin{align}
& \mathcal{U}^{min}_{ef}(r) = \mathcal{U}_{ef}(r^+_*) = -\frac{1}{2} 0.04602372915256611, \\
& \mathcal{U}^{max}_{ef}(r) = \mathcal{U}_{ef}(r^-_*) = \frac{1}{2} 0.2682459513747886.
\end{align}

#### Equação para órbitas

Dado $\frac{dr}{d\lambda} = \frac{dr}{d\phi} \frac{d\phi }{d\lambda}= \frac{dr}{d\phi} \frac{L }{r^2}$ e definindo $u = m/r$, obtém-se que
\begin{align}
(\frac{du}{d\phi})^2 = 2 u^3 - u^2 + 2 \frac{ m^2}{l^2} u + \frac{m^2}{l^2}(\epsilon^2 - 1),
\end{align}
que, derivando mais uma vez, obtemos
\begin{align}
\frac{d^2 u}{d\phi^2} = 3 u^2 - u +  \frac{ m^2}{l^2}.
\end{align}
Se fizermos $u = 1/r$, teremos $\frac{d^2 u}{d\phi^2} = 3 m u^2 - u +  \frac{ m}{l^2}$.

Seja agora
\begin{align}
y_0 = u, \ \ \ y_1 = \frac{d u }{d\phi} := u',
\end{align}
de modo que obtemos o seguinte sistema de EDO's
\begin{align}
& y'_1 = 3 y^2_0 - y_0 + \alpha, \ \ \ \ \alpha =  \frac{ m^2}{l^2},\\
& y'_0 = y_1,
\end{align}
cujos valores iniciais são
\begin{align}
y_0(\phi_0) = u(\phi_0) = u_0, \ \ \ \ \ y_1(\phi_0) = u'(\phi_0) = u'_0,
\end{align}
onde
\begin{align}
u(\phi_0) = u_0 = 1/r_0, \ \ \ \ u'(\phi_0) = u'_0 = -r'_0/r^2_0.
\end{align}
Mas dado que 
$$
\dot{r} = r' \dot{\phi} = r' (L/r^2) \implies r' = \frac{\dot{r} r^2}{L}.
$$
Ou seja,
\begin{align}
u(\phi_0) = u_0 = 1/r_0, \ \ \ \ u'(\phi_0) = u'_0 = -\dot{r}_0/L.
\end{align}


```python
#método RK4

lam=0.2372286066654316 #### alpha = 1/lam = m/l^2 = m/l^2c^2
rh=0.01976905055545263 #### rh = r_s

def fs(x,yvals):
    y0,y1=yvals
    f0=y1
    f1=(3*rh/2)*y0**2-y0+1/lam
    return np.array([f0, f1])

def rk4_gen(fs,x0,xf,n,yinits):
    h=(xf-x0)/(n-1)
    xs=x0+np.arange(n)*h
    ys=np.zeros((n, yinits.size))
    yvals=np.copy(yinits)
    for j,x in enumerate(xs):
        ys[j,:]=yvals
        k0=h*fs(x, yvals)
        k1=h*fs(x+h/2, yvals+k0/2)
        k2=h*fs(x+h/2, yvals+k1/2)
        k3=h*fs(x+h, yvals+k2)
        yvals += (k0 + 2*k1 + 2*k2 + k3)/6
    return xs, ys
```

\frac{d^2 u}{d\phi^2} = 3 u^2 - u,
\end{align}
ou ainda, $\frac{d^2 u}{d\phi^2} = 3m u^2 - u $.

```python
rh=0.01976905055545263

def fs(x,yvals):
    y0,y1=yvals
    f0=y1
    f1=(3*rh/2)*y0**2-y0
    return np.array([f0, f1])

def rk4_gen(fs,x0,xf,n,yinits):
    h=(xf-x0)/(n-1)
    xs=x0+np.arange(n)*h
    ys=np.zeros((n, yinits.size))
    yvals=np.copy(yinits)
    for j,x in enumerate(xs):
        ys[j,:]=yvals
        k0=h*fs(x, yvals)
        k1=h*fs(x+h/2, yvals+k0/2)
        k2=h*fs(x+h/2, yvals+k1/2)
        k3=h*fs(x+h, yvals+k2)
        yvals += (k0 + 2*k1 + 2*k2 + k3)/6
    return xs, ys
```

#### a) E = E_1

```python
rh=0.01976905055545263
rstar=0.02965357583317894
Vmax=6.046895227462673*10**(21)
c=63197.8
r0=10**20
r_dot0=np.sqrt(float(10**21))

#### Initial conditions required for RK4####
u0=1/r0
u_prime0=-r_dot0/(c**2)
############################################

#### Data obtained from RK4 #####
x0,xf,n=1.1859*np.pi,4*np.pi,1000
yinits = np.array([u0,u_prime0]) 
xs, ys = rk4_gen(fs,x0,xf,n,yinits)

y_s=np.zeros(n)
for i in range(len(ys[:,0])):
    y_s[i]=1/ys[i,0]
###############################
    
########## Plot ##########
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(xs,y_s)
ax.set_ylim(0,0.3)
ax.set_xlabel('$r\,\,(AU$)')

circle = plt.Circle((0,0), rh, transform=ax.transData._b, fc='gray',ec="k",alpha=0.4)
ax.add_artist(circle)

ax.grid(True)
ax.set_rticks([rh,0.2])
ax.set_rlabel_position(45)
rtick_labels = ['$r_h$','$0.2$']
ax.set_yticklabels(rtick_labels)
j=rtick_labels.index('$r_h$') 
rgrid=ax.yaxis.get_gridlines()
rgrid[j].set_linewidth(1)
rgrid[j].set_color('k')
plt.tight_layout()
```
